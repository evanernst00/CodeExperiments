# Matplotlib

### Basic graph experiments

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Set the style for plots
plt.style.use("classic")
```

```python
x = np.linspace(1, 10, 10)
print(x)

plt.plot(x, x) # x, y

plt.show()
```

```python
x = np.linspace(0, 10, 11)
print(x)

plt.plot(x, np.sin(x))

plt.show()
```

```python
x = np.linspace(0, 10, 101)
print(x)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```
