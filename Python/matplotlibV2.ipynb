# Matplotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Two Different Interfaces

### MATLAB Style Interface (pyplot/plt)
The first interface is a MATLAB-style state-based interface. It is more convenient but less powerful, and contained in pyplot (plt) interface.
There is a current, hidden figure and axes, where everything is applied, rather than making these explicit.

```python
x = np.linspace(0, 10, 100)
```

```python
# MATLAB Style Interface

plt.figure()  # create a plot figure

# Subplot 1
plt.subplot(2, 1, 1)  # (rows, columns, panel number)
plt.plot(x, np.sin(x), '-')
# Subplot 2
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x), '--')
```

### Object-Oriented Interface
The second interface is object-oriented, and more powerful
In this interface, the plotting functions are *methods* of Figure and Axes objects

```python
# Object-Oriented Interface

# Previously used "plt.subplot", which works on the CURRENT figure
# Now using "plt.subplots", which creates an independent object to operate on
fig, axs = plt.subplots(2)

# Access target object within the array, and operate on it
# Instead of accessing it within the active structure
axs[0].plot(x, np.sin(x), '-')
axs[1].plot(x, np.cos(x), '--')
```

```python
# Not limited to single interface
fig2, axs2 = plt.subplots(2, 2)
axs2[0, 0].plot(x, np.sin(x))

# also possible to create the figure and axes object separately
```

## Simple Plotting

```python
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)

# ax.legend() creates a legend using specified labels (in ax.plot())
ax.plot(x, np.sin(x - 0), '-b', label='Sine - 0')
ax.plot(x, np.sin(x - 1), '-g', label='Sine - 1')
ax.plot(x, np.sin(x - 2), label='Sine - 2', color='red')
ax.plot(x, np.sin(x - 3), label='Sine - 3', color='.4') # 0 to 1 color value (greyscale)

ax.legend()

```
