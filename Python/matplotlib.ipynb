# Matplotlib

## Basic graphing & styling

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Set the style for plots
plt.style.use("classic")
```

```python
x1 = [0, 10] # f(x) = x
x2 = np.linspace(0, 10, 6) # rough sine
x3 = np.linspace(0, 10, 51) # smooth sine
print("Linear:", x1)
print("Sin rough:", x2)
print("Sin smooth:", x3)

plt.plot(x1, x1) # x, f(x)
plt.plot(x3, np.sin(x3), "bo") # smooth sine
plt.plot(x2, np.sin(x2)) # rough sine

# Modify the graph, add labels
plt.title("Basic Graphing", fontdict={"fontname":"Roboto", "fontsize":25})
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.show()
```
